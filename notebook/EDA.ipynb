```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#distribution of the target feature name 'Class'
plt.figure(figsize=(6,4))
sns.countplot(x="Class",data=df)
plt.title("Distribution of the fraud and non-fraud transactions")
```

this data is imblance ,we have 492 frauds out of 284,807 transactions.The dataset is highly unbalanced, the positive class (frauds =1) account for 0.172% of all transactions.

```python
# Distribution of 'Amount' feature
plt.figure(figsize=(8, 6))
sns.histplot(df['Amount'], bins=30, kde=True)
plt.title('Distribution of Transaction Amounts')
plt.show()
```

```python
#Distribution of 'Time' feature
plt.figure(figsize=(8, 6))
sns.histplot(df['Time'], bins=50, kde=True)
plt.title('Distribution of Transaction Times')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12, 10))
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm', annot=False, fmt='.2f')
plt.title('Correlation Matrix of Features')
plt.show()
```

```python
# Boxplot to check outliers in the 'Amount' feature
plt.figure(figsize=(8, 6))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title('Boxplot of Transaction Amounts by Class')
plt.show()
```

```python
# Boxplot to check outliers in the 'Time' feature
plt.figure(figsize=(8, 6))
sns.boxplot(x='Class', y='Time', data=df)
plt.title('Boxplot of Transaction Times by Class')
plt.show()
```

```python
# Distribution of principal components (V1 to V28)
plt.figure(figsize=(14, 10))
for i in range(1, 29):
    plt.subplot(7, 4, i)
    sns.histplot(df[f'V{i}'], bins=30, kde=True)
    plt.title(f'Distribution of V{i}')
plt.tight_layout()
plt.show()
```

```python
# Pairplot for a subset of features including 'Amount' and 'Time'
subset = df[['V1', 'V2', 'V3', 'Amount', 'Time', 'Class']].sample(5000, random_state=1)
sns.pairplot(subset, hue='Class', diag_kind='kde')
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()
```
